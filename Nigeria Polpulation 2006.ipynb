### 1.1 Nigeria Total, Average Population and Deviation as at 2006

```python
print("The Total Population in Nigeria as at 2006 was ", pop['Population'].sum())
print("------------------------------------------------------------------")
print("The Average Population in Nigeria as at 2006 was ", int(pop['Population'].mean()))
print("------------------------------------------------------------------")
print("The Total Population Deviation in Nigeria as at 2006 was ", int(pop['Population'].std()))
```

### 1.2 State with the Highest population and the population

```python
max_population = pop["Population"].max()
# State with high population =HPS
HPS = [g for g, x in enumerate(pop["Population"]) if x== max_population]

#iterate through the population and get the state with the higest population
for g in HPS:
    print(pop["STATES"][g],"with a population of", pop["Population"][g],"has the Highest Population in Nigeria" )
```

```python
min_population = pop["Population"].min()
# State with lowest population =lPS
LPS = [g for g, x in enumerate(pop["Population"]) if x== min_population]

#iterate through the population and get the state with the higest population
for g in LPS:
    print(pop["STATES"][g],"with a population of", pop["Population"][g],"has the Lowest Population in Nigeria" )
```

### 1.3 State which population is greater than average population in Nigeria as at 2006

```python
avg_population = pop["Population"].mean()
# State with high population =APS
APS = [g for g, x in enumerate(pop["Population"]) if x>= avg_population]

#iterate through the population ang get the states that the populations are above average
for g in APS:
    print(pop["STATES"][g],"with a population of", pop["Population"][g],"has a population higer than the average population \n------------" )
```

Fouteen state have a population greater than the average populations

```python
print ("Nigeria population from top top to less ",pop.sort_values("Population", axis = 0, ascending = False))
```

```python
plt.figure(figsize =(15,7))
total_population_by_region = [SE['population'].sum(),SW['population'].sum(),SS['population'].sum(),NE['population'].sum(),
                             NW['population'].sum(),NC['population'].sum()]
Regions = ['SE', 'SW', 'SS', 'NE', 'NW', 'NC']
total_population_by_region2 =[round(x/1000000) for x in total_population_by_region]


    

clr =['green','blue','pink','yellow','red']
plt.barh(Regions, total_population_by_region2, color= clr)
plt.xlabel('Population in millions')
plt.ylabel('Regions')
plt.title("population by Region  ")
plt.style.use('seaborn-whitegrid') 
```

### BY LAND AREA (km2)

```python
plt.figure(figsize =(15,7))
total_area_by_region = [SE['area'].sum(),SW['area'].sum(),SS['area'].sum(),NE['area'].sum(),
                             NW['area'].sum(),NC['area'].sum()]
Regions = ['SE', 'SW', 'SS', 'NE', 'NW', 'NC']
plt.bar(Regions, total_area_by_region, color='g', width = 0.5)
plt.xlabel('Regions')
plt.ylabel('Land Area(km2)')
plt.title("Land mass by Region  ")
plt.style.use('seaborn-whitegrid') 
```
